This dataset is unbalanced, thus we will have to handle this.  
  0 --> Normal Transaction  
  1 --> Fraudulent Transaction  

```python
legit = df[df.Class==0]
fraud = df[df.Class==1]
legit.shape, fraud.shape
```

```python
df.groupby('Class').mean()
```

## Perform Undersampling  
Build a dataset containing similar transaction distribution

```python
legit_sample = legit.sample(n = 492)
```

```python
legit_sample.head()
```

```python
legit_sample.shape
```

```python
# Concatinating legit_sample and fraudelent data frame
```

```python
df_sample = pd.concat([legit_sample, fraud], axis =0)
```

```python
df_sample.head()
```

```python
df_sample.tail()
```

```python
df_sample.Class.value_counts()
```

```python
df_sample.groupby("Class").mean()
```

## Dataset Split

```python
X = df_sample.drop(columns='Class', axis = 1)
Y = df_sample['Class']
X,Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify = Y, random_state=2)
```

Model Training

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

Model Evaluation

```python
X_train_pred = model.predict(X_train)
train_accuracy = accuracy_score(X_train_pred, Y_train)
train_accuracy
```

```python
X_test_pred = model.predict(X_test)
test_accuracy = accuracy_score(X_test_pred, Y_test)
test_accuracy
```
